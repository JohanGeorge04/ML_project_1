```python
from sklearn.model_selection import train_test_split

df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=11)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=11)
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

y_train = (df_train.diagnosed_diabetes).astype('int').values
y_val = (df_val.diagnosed_diabetes).astype('int').values
y_test = (df_test.diagnosed_diabetes).astype('int').values

del df_train['diagnosed_diabetes']
del df_val['diagnosed_diabetes']
del df_test['diagnosed_diabetes']
```

## Decision Tree Classifier


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_extraction import DictVectorizer
from sklearn.metrics import roc_auc_score

from sklearn.tree import export_text
```

```python
train_dicts = df_train.to_dict(orient='records')

dv = DictVectorizer(sparse=False)
X_train = dv.fit_transform(train_dicts)

dt = DecisionTreeClassifier(max_depth=6)
dt.fit(X_train, y_train)

val_dicts = df_val.to_dict(orient='records')
X_val = dv.transform(val_dicts)

y_pred = dt.predict_proba(X_val)[:, 1]
roc_auc_score(y_val, y_pred)
```

```python
print(export_text(dt, feature_names=list(dv.get_feature_names_out())))
```

```python
scores = []

for depth in [2,4, 5, 6,10]:
    for s in [1, 5, 10, 15, 20, 500, 100, 200]:
        dt = DecisionTreeClassifier(max_depth=depth, min_samples_leaf=s)
        dt.fit(X_train, y_train)

        y_pred = dt.predict_proba(X_val)[:, 1]
        auc = roc_auc_score(y_val, y_pred)
        
        scores.append((depth, s, auc))
```

```python
columns = ['max_depth', 'min_samples_leaf', 'auc']
df_scores = pd.DataFrame(scores, columns=columns)
```

```python
df_scores_pivot = df_scores.pivot(index='min_samples_leaf', columns=['max_depth'], values=['auc'])
df_scores_pivot
```

```python
full_train_dicts = df_full_train.to_dict(orient='records')

dv = DictVectorizer(sparse=False)
X_full_train = dv.fit_transform(full_train_dicts)

dt = DecisionTreeClassifier(max_depth=5)
y_full_train= np.concatenate([y_train, y_val])
dt.fit(X_full_train, y_full_train)

test_dicts = df_test.to_dict(orient='records')
X_test = dv.transform(test_dicts)

y_pred = dt.predict_proba(X_test)[:, 1]
roc_auc_score(y_test, y_pred)
```

## RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
scores = []

for d in [5, 10, 15]:
    for n in range(10, 101, 10):
        rf = RandomForestClassifier(n_estimators=n,
                                    max_depth=d,
                                    random_state=1)
        rf.fit(X_train, y_train)

        y_pred = rf.predict_proba(X_val)[:, 1]
        auc = roc_auc_score(y_val, y_pred)

        scores.append((d, n, auc))
```

```python
columns = ['max_depth', 'n_estimators', 'auc']
df_scores = pd.DataFrame(scores, columns=columns)
```

```python
for d in [5, 10, 15]:
    df_subset = df_scores[df_scores.max_depth == d]
    
    plt.plot(df_subset.n_estimators, df_subset.auc,
             label='max_depth=%d' % d)

plt.legend()
```

```python
max_depth = 10
scores = []

for s in [1, 3, 5, 10, 50]:
    for n in range(10, 101, 10):
        rf = RandomForestClassifier(n_estimators=n,
                                    max_depth=max_depth,
                                    min_samples_leaf=s,
                                    random_state=1)
        rf.fit(X_train, y_train)

        y_pred = rf.predict_proba(X_val)[:, 1]
        auc = roc_auc_score(y_val, y_pred)

        scores.append((s, n, auc))
```

```python
columns = ['min_samples_leaf', 'n_estimators', 'auc']
df_scores = pd.DataFrame(scores, columns=columns)
```

```python
colors = ['black', 'blue', 'orange', 'red', 'grey']
values = [1, 3, 5, 10, 50]

for s, col in zip(values, colors):
    df_subset = df_scores[df_scores.min_samples_leaf == s]
    
    plt.plot(df_subset.n_estimators, df_subset.auc,
             color=col,
             label='min_samples_leaf=%d' % s)

plt.legend()
```

## XG Boost


```python
import xgboost as xgb
```

```python
features = list(dv.get_feature_names_out())
dtrain = xgb.DMatrix(X_train, label=y_train, feature_names=features)
dval = xgb.DMatrix(X_val, feature_names=features)
```

```python
xgb_params = {
    'eta': 0.3, 
    'max_depth': 10,
    'min_child_weight': 1,
    
    'objective': 'binary:logistic',
    'nthread': 8,
    
    'seed': 1,
    'verbosity': 1,
}

model = xgb.train(xgb_params, dtrain, num_boost_round=10)
```

```python

y_pred = model.predict(dval)
y_pred
```

```python
roc_auc_score(y_val, y_pred)
```

## Cross-Validation

```python
kfold = KFold(n_splits=10, shuffle=True, random_state=1)

for train_idx, val_idx in kfold.split(df_full_train):
        df_train = df_full_train.iloc[train_idx]
        df_val = df_full_train.iloc[val_idx]

        y_train = (df_train.diagnosed_diabetes).astype('int').values
        y_val = (df_val.diagnosed_diabetes).astype('int').values
```

```python
train_dicts = df_train.to_dict(orient='records')

dv = DictVectorizer(sparse=False)
X_train = dv.fit_transform(train_dicts)

dt = DecisionTreeClassifier(max_depth=6)
dt.fit(X_train, y_train)

val_dicts = df_val.to_dict(orient='records')
X_val = dv.transform(val_dicts)

y_pred = dt.predict_proba(X_val)[:, 1]
roc_auc_score(y_val, y_pred)
```

```python

test_dicts = df_test.to_dict(orient='records')
X_test = dv.transform(test_dicts)

y_pred = dt.predict_proba(X_test)[:, 1]
roc_auc_score(y_test, y_pred)
```
